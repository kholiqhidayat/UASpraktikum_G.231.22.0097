**Membuat dataset dengan saya beri nama "data" yang berisi file stock.csv**

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("stock1.csv")
```

```python
plt.hist(data);
```

**Distribusi data pada setiap atribut Purchase Price**

```python
# Histogram kedua
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(data['Purchase Price'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Purchase Price',fontsize=15, color='red')
plt.xlabel('Purchase Price', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

**Distribusi data pada setiap atribut Current Price**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.hist(data['Current Price'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Current Price',fontsize=15, color='red')
plt.xlabel('Current Price', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

**Distribusi data pada setiap atribut Gain/Loss**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.hist(data['Gain/Loss'], range=(1,12))
plt.title('Distribusi data pada setiap atribut Gain/Loss',fontsize=15, color='red')
plt.xlabel('Gain/Loss', fontsize = 12)
plt.ylabel('Jumlah Banyaknya', fontsize = 12)
plt.xlim(xmin=0,xmax=12)
plt.show()
```

**Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
Dividend_share = (data.groupby('Stock Name')['Dividend/share']
                  .sum()
                  .reset_index()
                  .sort_values(by='Dividend/share', ascending=False)
                  .head(5))
print(Dividend_share)
```
